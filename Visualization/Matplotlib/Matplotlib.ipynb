```python
x = np.arange(1,6)
y = np.arange(2,11,2)

#plt.plot(x,y,"red")
fig = plt.figure() 
axes1 = fig.add_axes([0.1,0.2,0.8,0.9])
axes2 = fig.add_axes([0.2,0.8,0.3,0.2])

axes1.plot(y,x)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(y,x)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")
plt.show()
```

```python
fig,axes = plt.subplots(nrows= 2, ncols=1)
# ---> nrows ve ncols değerleri ile oyanayarak mantığı daha iyi kavra.
for ax in axes:
    ax.plot(x,y) # her grafiğe aynı şey çizilecekse
    '''
axes[0].plot(x,y)
axes[0].set_title("First Axes")
axes[1].plot(x,x**3) 
axes[1].set_title("Second Axes")
---------> her sub kısmına farklı grafikler çizilecekse indexleri birer birer özelleştirirz.
'''
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(51)
y = np.arange(0,101,2)
z = x ** 3
```

```python
fig = plt.figure(figsize=(6,5))
axes = fig.add_axes([0,0,1,1])
plt.grid()

axes.plot(x,y,color="red")
axes.set_xlabel("X values")
axes.set_ylabel("Y values")
axes.set_title("X ve Y Graph")
```

```python
''' 1. Figure Size'ı (8,6) olan bir figure Oluşturun
    2. axes1 ve axes2 adında 2 adet axes oluşturun.(axes1 pozisyonu = [0,0.1,0.8,0.8], axes2 
    pozisyonu = [0.1,0.5,0.4,0.3])
    3. axes1 üzerinde x vs z grafiğini, axes2 üzerinde x vs y grafiğini çizin ve aşağıdaki 
    grafiğe benzer bir grafik oluşturun. '''

fig = plt.figure(figsize=(8,6)) 
axes1 = fig.add_axes([0,0.1,0.8,0.8])
axes1.plot(x,x**3,color="purple")
axes1.set_xlabel("X values")
axes1.set_ylabel("Y values")
axes1.set_title("Outer Graph X VE Z")


axes2 = fig.add_axes([0.1,0.5,0.4,0.3])
axes2.plot(x,y,color="red")
axes2.set_xlabel("X values")
axes2.set_ylabel("Z values")
axes2.set_title("Inner Graph X VE Y")
```

```python
'''1. 2 adet yanyana subplot oluşturun. (figsize=(10,6))
2. Birinci grafiğe (x,y) grafiğini , ikinci grafiğe (x,z) grafiğini çizin.
3. plot() fonksiyonunun parametrelerini vererek, grafiği aşağıdaki grafiğe benzer özelleştirin.'''

fig,axes = plt.subplots(nrows= 1, ncols=2,figsize = (7,5))

axes[0].plot(x,y,color="red",lw=1,ls="--",marker="o",
         markerfacecolor="black",markeredgecolor="red",markeredgewidth=1)
axes[0].set_xlabel("X values")
axes[0].set_ylabel("Y values")
axes[0].set_title("X ve Y Graph")

axes[1].plot(x,z,color="purple",lw=3,ls="-.")
axes[1].set_xlabel("X values")
axes[1].set_ylabel("Z values")
axes[1].set_title("X ve Z Graph")

plt.tight_layout()
```
